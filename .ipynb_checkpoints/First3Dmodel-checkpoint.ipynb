# Run basic 3D convolutional network and make some tests.

## Import and normalize Dataset.

```python
import numpy as np
import random
```

```python
CancerDataset = np.load('CancerPatients.npy')
HealthyDataset = np.load('HealthyPatients.npy')

CancerDataset.shape, HealthyDataset.shape
```

```python
CancerDataset = CancerDataset/CancerDataset.max()

cancer_std = np.std(CancerDataset)
cancer_mean = np.mean(CancerDataset)

CancerDataset = CancerDataset - cancer_mean
CancerDataset = CancerDataset/cancer_std
```

```python
HealthyDataset = HealthyDataset/HealthyDataset.max()

healthy_std = np.std(HealthyDataset)
healthy_mean = np.mean(HealthyDataset)

HealthyDataset = HealthyDataset - healthy_mean
HealthyDataset = HealthyDataset/healthy_std
```

```python
np.mean(CancerDataset), np.std(CancerDataset), np.mean(HealthyDataset), np.std(HealthyDataset)
```

### Combine Datasets.

```python
Dataset = np.concatenate((CancerDataset, HealthyDataset), axis=0)

Dataset.shape
```

```python
Cancer_Labels = [[0,1]] * 110
Healthy_Labels = [[1,0]] * 110

len(Cancer_Labels), len(Healthy_Labels)
```

```python
Cancer_Labels = np.array(Cancer_Labels)
Healthy_Labels = np.array(Healthy_Labels)

labels = np.concatenate((Cancer_Labels, Healthy_Labels))
labels.shape
```

```python
Test_Cancer_Dataset = Dataset[:5]
Test_Healthy_Dataset = Dataset[-5:]
```

```python
Dataset = Dataset[5:-5]
labels = labels[5:-5]

len(Dataset), len(labels)
```

```python
X = Dataset
y = labels
```

```python
shuffle_list = list(zip(X,y))
```

```python
random.shuffle(shuffle_list)
```

```python
X, y = zip(*shuffle_list)
```

```python
X = np.array(X).reshape(210,128,128,17,1)
y = np.array(y)

X.shape, y.shape
```

## Create basic model.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv3D, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import Accuracy, Recall, Precision
from tensorflow.keras.losses import CategoricalCrossentropy
```

```python
model = Sequential()

# Input 128x128x17x1
model.add(Conv3D(4, kernel_size=(3, 3, 3), strides=(2, 2, 2), kernel_initializer='he_uniform', input_shape=(128, 128, 17, 1)))
model.add(BatchNormalization(center=True, scale=True))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# Input 64x64x9x4
model.add(Conv3D(8, kernel_size=(3, 3, 3), strides=(2, 2, 2), kernel_initializer='he_uniform', input_shape=Dataset.shape[1:]))
model.add(BatchNormalization(center=True, scale=True))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# Input 32x32x5x8
model.add(Conv3D(16, kernel_size=(3, 3, 3), strides=(2, 2, 2), kernel_initializer='he_uniform', input_shape=Dataset.shape[1:]))
model.add(BatchNormalization(center=True, scale=True))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# Input 16x16x3x16
model.add(Flatten())
model.add(Dense(256, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(128, activation='relu', kernel_initializer='he_uniform'))

model.add(Dense(2, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy', Recall(), Precision()])
```

```python
model.fit(X, y, batch_size=2, epochs=20, validation_split=0.3)
```

```python
model.summary()
```

## Build heatmap

```python
import cv2
import matplotlib.pyplot as plt

import tensorflow.keras.backend as K
```

```python
def PlotHeatMapOverSlice(Dataset, patient_index, patient_slice, intensity=0.5):
    
    choosen_slice = Dataset[patient_index,:,:,patient_slice]
    choosen_slice = np.squeeze(choosen_slice, axis=2)
    plt.imshow(choosen_slice, cmap='gray')
    plt.show()
    
    patient = Dataset[patient_index:patient_index+1]
    
    with tf.GradientTape() as tape:
        last_conv_layer = model.get_layer('conv3d_2')
        iterate = Model([model.inputs], [model.output, last_conv_layer.output])
        model_out, last_conv_layer = iterate(patient)
        class_out = model_out[:, np.argmax(model_out[0])]
        grads = tape.gradient(class_out, last_conv_layer)
        pooled_grads = K.mean(grads, axis=(0,1,2,3))

    heatmap = tf.reduce_mean(tf.multiply(pooled_grads, last_conv_layer), axis=-1)
    
    heatmap = np.maximum(heatmap, 0)  
    if not np.max(heatmap) == 0:
        heatmap /= np.max(heatmap)
    heatmap = heatmap.reshape((15, 15))
    
    patient_slice_img = patient[0,:,:,patient_slice]
    patient_slice_img = np.concatenate((patient_slice_img, 
                                        patient_slice_img, patient_slice_img), axis=2)
    patient_slice_img = patient_slice_img - np.min(patient_slice_img)
    patient_slice_img = patient_slice_img * (254/np.max(patient_slice_img))
    
    heatmap = cv2.resize(heatmap, (patient_slice_img.shape[:2]))
    heatmap = cv2.applyColorMap(np.uint8(255*heatmap), cv2.COLORMAP_JET)
    
    plt.imshow(heatmap)
    plt.show()
    
    
    image = heatmap * intensity + patient_slice_img
    image = image/np.max(image)
    
    plt.imshow(image)
    plt.show()
```

```python
PlotHeatMapOverSlice(Test_Cancer_Dataset, 0, 11)
```

```python
PlotHeatMapOverSlice(Test_Cancer_Dataset, 3, 10)
```

```python
PlotHeatMapOverSlice(Test_Cancer_Dataset, 0, 10)
```

```python
PlotHeatMapOverSlice(Test_Healthy_Dataset, 3, 10)
```

```python
model = Sequential()

# Input 128x128x17x1
model.add(Conv3D(4, kernel_size=(3, 3, 3), strides=(2, 2, 2), kernel_initializer='he_uniform', input_shape=(100, 100, 17, 1)))
model.add(BatchNormalization(center=True, scale=True))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# Input 64x64x9x4
model.add(Conv3D(8, kernel_size=(3, 3, 3), strides=(2, 2, 2), kernel_initializer='he_uniform', input_shape=Dataset.shape[1:]))
model.add(BatchNormalization(center=True, scale=True))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# Input 32x32x5x8
model.add(Conv3D(16, kernel_size=(3, 3, 3), strides=(2, 2, 2), kernel_initializer='he_uniform', input_shape=Dataset.shape[1:]))
model.add(BatchNormalization(center=True, scale=True))
model.add(Activation('relu'))
model.add(Dropout(0.5))

# Input 16x16x3x16
model.add(Flatten())
model.add(Dense(256, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(128, activation='relu', kernel_initializer='he_uniform'))

model.add(Dense(2, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy', Recall(), Precision()])
```

```python
model.fit(Dataset_DA, y, batch_size=2, epochs=20, validation_split=0.3)
```
