# Testing ANTsPy for brain MR images normalization

### Check the structure and quality of the downloaded dataset of cancer patients sampled from TCGA-GBM.
### Remember that images are in dicom format.

```python
import os
import random
import glob
import ants
import nibabel as nib
import pydicom
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
os.getcwd()
```

```python
patients_folder = Path('../Datasets//Sample_TCGA-GBM/TCGA-GBM/')

patients_ids = os.listdir(patients_folder)
patients_ids, len(patients_ids)
```

### Inside each patient folder, there are two more directories before dicom files.
### Let's visualize the intermediate slice of two random patients and check its format. Patients with differents 3D scan methods will be chosen

```python
p1 = random.choice(patients_ids)
p2 = random.choice(patients_ids)

p1_scan_path = glob.glob(os.path.join(patients_folder, p1, '*', '*'))[0]
p2_scan_path = glob.glob(os.path.join(patients_folder, p2, '*', '*'))[0]

p1_scan_path, p2_scan_path
```

```python
p1_slice = sorted(os.listdir(p1_scan_path))[len(p1_scan_path)//2]
p2_slice = sorted(os.listdir(p2_scan_path))[len(p2_scan_path)//2]

p1_slice_path = os.path.join(p1_scan_path, p1_slice)
p2_slice_path = os.path.join(p2_scan_path, p2_slice)

p1_slice_path, p2_slice_path
```

## ANTsPy library reading dicom file. It's observed that the dicom information is not read.

```python
p1_img = ants.image_read(p1_slice_path, dimension=2)
p2_img = ants.image_read(p2_slice_path, dimension=2)

p1_img, p2_img
```

```python
ants.plot(p1_img)
ants.plot(p2_img)
```

```python
ants.plot(p1_img, overlay=p2_img, cmap='gray', overlay_cmap='gray', overlay_alpha=0.4)
```

```python
normalized = ants.registration(fixed=p1_img, moving=p2_img, type_of_transform="SyN")

normalized
```

```python
ants.plot(normalized['warpedfixout'], cmap='gray', overlay=normalized['warpedmovout'], overlay_cmap='gray', overlay_alpha=0.5)
```

## Plotting images from different datasets and with different number of dimensions is not being possible. The last attempt will be to read two copies of each dataset from numpy arrays.

### Let's construct a 3D dicom image file of patient P1.

```python
p1_slices = os.listdir(p1_scan_path)
p1_list = []

for curr_slice in p1_slices:
    slice_path = os.path.join(p1_scan_path, curr_slice)
    dicom_data = pydicom.dcmread(slice_path)
    slice_array = np.expand_dims(dicom_data.pixel_array, axis=2)
    p1_list.append(slice_array)

p1_3d = np.dstack(p1_list)
p1_3d = p1_3d.astype('float32')
p1_3d.shape, p1_3d.min(), p1_3d.max()
```

```python
p1_array_ants = ants.from_numpy(p1_3d)
p1_array_ants
```

### Recall that hp_object is a nibabel object and hp_array is its 3d array.
### Check is max values is 1.0 as well.

```python
hp_array.shape, hp_array.min(), hp_array.max()
```

```python
hp_array_ants = ants.from_numpy(hp_array)
hp_array_ants
```

### It's crucial to remember that the dimension with less slices (different from 256) are not the same for the two datasets. The plots below will illustrate clearly.
### Not even the axial axis is the same.

```python
ants.plot(p1_array_ants, axis=2, slices=(80), figsize=3)

ants.plot(hp_array_ants, axis=1, slices=(150), figsize=2)
```

### So let's change the axes. The depth convention on axis 2 will be adopted.

```python
hp_array_T = hp_array.transpose(0,2,1)
hp_array_T.shape
```

```python
hp_array_T_ants = ants.from_numpy(hp_array_T)
hp_array_T_ants
```

```python
ants.plot(p1_array_ants, axis=2, slices=(80), figsize=3)

ants.plot(hp_array_T_ants, axis=2, slices=(150), figsize=2)
```
